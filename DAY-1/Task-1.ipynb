```python
# Importing necessary libraries
import pandas as pd

# Loading the dataset
netflix_data = pd.read_csv('netflix_titles.csv', encoding='utf-8')

# Displaying the first few rows of the dataset to understand its structure
netflix_data.head()
```

```python
# Checking for missing values in the dataset
missing_values = netflix_data.isnull().sum()

# Displaying the missing values
missing_values
```

```python
# Dropping rows with missing values in critical columns
netflix_data_cleaned = netflix_data.dropna(subset=['director', 'cast', 'country', 'date_added', 'rating', 'duration'])

# Checking the shape of the cleaned dataset
cleaned_shape = netflix_data_cleaned.shape

# Displaying the first few rows of the cleaned dataset
cleaned_head = netflix_data_cleaned.head()
```

```python
# Displaying the shape of the cleaned dataset and the first few rows
cleaned_shape, cleaned_head
```
